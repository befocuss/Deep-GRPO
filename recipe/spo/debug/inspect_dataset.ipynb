## Dr.GRPO

```python
import datasets
import os
```

```python
DATA_PATH = "/root/paddlejob/workspace/understand-r1-zero/datasets"
```

```python
from typing import List, Optional
```

```python
def check_equivalent_columns(
    dataset: datasets.Dataset,
    columns: List[str],
    sample_size: Optional[int] = None,
    ignore_nan: bool = True
) -> bool:
    """
    检查数据集中的若干列是否等价
    
    Args:
        dataset: HuggingFace数据集
        columns: 要检查的列名列表
        sample_size: 检查的样本数量，None表示检查全部
        ignore_nan: 是否忽略NaN值的差异
    
    Returns:
        bool: 所有列是否等价
    """
    if len(columns) < 2:
        raise ValueError("至少需要提供两个列名")
    
    # 检查列是否存在
    for col in columns:
        if col not in dataset.column_names:
            raise ValueError(f"列 '{col}' 不存在于数据集中")
    
    # 获取要检查的数据
    if sample_size is not None:
        data_subset = dataset.select(range(min(sample_size, len(dataset))))
    else:
        data_subset = dataset
    
    # 检查每一行是否所有列都相等
    for i in range(len(data_subset)):
        row_values = []
        for col in columns:
            value = data_subset[i][col]
            
            if ignore_nan and isinstance(value, float):
                # 处理NaN值
                if value != value:  # NaN检查
                    value = float('nan')
            
            row_values.append(value)
        
        # 检查当前行中所有列的值是否相等
        if len(set(row_values)) > 1:
            # 处理特殊情况：所有NaN应该被视为相等
            if ignore_nan and all(isinstance(v, float) and v != v for v in row_values):
                continue
            return False
    
    return True
```
